# 1. Python Codes - Basics

## 1.1. Numbers

### 1.1.1. Simple Calculator
The interpreter acts as a simple calculator: we can type an expression into it and it will write the value. Expression syntax is straightforward: the operators +, -, * and / can be used to perform arithmetic; parentheses (()) can be used for grouping. For example:

```python
2 + 2
```

```python
50 - 5*6
```

```python
(50 - 5*6)/4
```

```python
8 / 5 # division always returns a floating-point number
```

### 1.1.2. Floor Division
The integer numbers (e.g. 2, 4, 20) have type int, the ones with a fractional part (e.g. 5.0, 1.6) have type float. We will see more about numeric types later in the tutorial.

Division (/) always returns a float. To do floor division and get an integer result we can use the // operator; to calculate the remainder you can use %:

```python
17 / 3 # classic division returns a float
```

```python
17 // 3  # floor division discards the fractional part
```

```python
17 % 3  # the % operator returns the remainder of the division
```

```python
5 * 3 + 2 # floored quotient * divisor + remainder
```

### 1.1.3. Calculate Powers
With Python, it is possible to use the ** operator to calculate powers :

```python
5 ** 2  # 5 squared
```

```python
2 ** 10  # 2 to the power of 7
```

### 1.1.4. Assign Value
* The equal sign (=) is used to assign a value to a variable. Afterwards, no result is displayed before the next interactive prompt:

```python
width = 20
height = 5 * 9
width * height
```

* If a variable is not “defined” (assigned a value), trying to use it will give you an error:

```python
a
```

### 1.1.5. Use of '_'
In interactive mode, the last printed expression is assigned to the variable _. This means that when we are using Python as a desk calculator, it is somewhat easier to continue calculations, for example:

```python
tax = 12.5 / 100
price = 100.50
price * tax
```

```python
price + _
```

```python
round(_, 2)
```
