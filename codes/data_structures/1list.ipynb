## Data Structures

### 1. List

```python
fruits = ['orange', 'apple', 'pear', 'banana', 'kiwi', 'apple', 'banana']

print(fruits.count('apple'))
print(fruits.count('tangerine'))

print(fruits.index('banana'))
print(fruits.index('banana', 4))

print(fruits.reverse())

fruits.append('grape')
print(fruits)

fruits.sort()
print(fruits)

print(fruits.pop())
```

### 2. Using Lists as Stacks

```python
stack = [3, 4, 5]
stack.append(6)
stack.append(7)
print("new stack ",stack)

stack.pop()
print(stack)

stack.pop()
stack.pop()
print("after final pop",stack)
```

### 3. Using Lists as Queues

```python
from collections import deque
queue = deque(['a', 'b', 'c'])
queue.append('d')
queue.append('e')
print("new queue ",queue)

print(queue.popleft())
print(queue.popleft())
print("Final queue: ",queue)
```

### 4. Create a list of squares

```python
# using normal function
square = []
def squar():
  for i in range(10):
    square.append(i**2)

squar()
print(square)
```

```python
# using lambda
square = list(map(lambda x: x**2, range(10)))
print(square)
```

```python
square = [x**2 for x in range(10)]
print(square)
```
